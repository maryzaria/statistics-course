```python
from urllib.request import urlopen, urlretrieve
from bs4 import BeautifulSoup
# urlretrieve(,'osm/map.osm')
import math

def getsqr(coordlist):
     baselat = coordlist[0][0]
     baselon = coordlist[0][1]
     degreelen = 111300
     newcoord = []
     for now in coordlist:
          newcoord.append(((now[0] - baselat) * degreelen, (now[1] - baselon) * degreelen * math.sin(baselat)))
     sqr = 0
     for i in range(len(newcoord) - 1):
          sqr += newcoord[i][0] * newcoord[i + 1][1] - newcoord[i + 1][0] * newcoord[i][1]
     sqr += newcoord[-1][0] * newcoord[0][1] - newcoord[0][0] * newcoord[-1][1]
     return abs(sqr)
    
resp = urlopen('file:///C:/Users/user/PycharmProjects/pythonProject/mapcity.osm') # скачиваем файл
xml = resp.read().decode('utf8')
soup = BeautifulSoup(xml)
id_koor = {}
for node in soup.find_all('node'):
    lat = float(node['lat'])
    lon = float(node['lon'])
    id_node = int(node['id'])
    id_koor[id_node] = (lat, lon)
dct={}
for way in soup.find_all('way'):
    id_way = int(way['id'])
    id_list = []
    for nd in way('nd'):
        ref = nd['ref']
        id_list.append(ref)
    if int(id_list[0]) == int(id_list[-1]):
        for tag in way('tag'):
            if tag['k'] == 'building':
                #print(id_way)
                lst=[]
                for i in range(len(id_list)):
                    a = int(id_list[i])
                    lst.append(id_koor[a])
                dct[id_way] = getsqr(lst)
biggest_S = sorted(dct.items(), key=lambda x: x[1], reverse=True)
print(biggest_S[0][0])
```
